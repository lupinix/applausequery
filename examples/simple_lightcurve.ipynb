# Simple example: Query a lightcurve from APPLAUSE using applausequery

We want to query the APPLAUSE DR3 lightcurve of star UCAC4 104-010297 and plot it. First of all we import the lightcurve submodule of applausequery.dr3 and matplotlib.pyplot.

```python
from applausequery.dr3 import lightcurve
import matplotlib.pyplot as plt 
%matplotlib inline
```

Now we query the lightcurve

```python
lc = lightcurve.lc_by_ucac4_id("104-010297")
plt.errorbar(lc["jd_mid"],lc["bmag"],yerr=lc["bmagerr"],fmt="o")
```
